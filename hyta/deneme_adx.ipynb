### SMOOTHED +DM14, -DM14, ATR
Smooth the 14-period averages of +DM, -DM, and the TR. Below is the formula for TR. <br>
Insert the -DM and +DM values to calculate the smoothed averages of those as well. <br>
First 14TR = Sum of first 14 TR readings. <br>
Next 14TR value = First 14TR - (Prior 14TR/14) + Current TR <br>

```python
def smoothed_plus_dm14():
    first_14_plus_dm = adx_data["+DM 1"].rolling(14).sum()
    adx_data["smoothed_plus_dm14"] = first_14_plus_dm - (first_14_plus_dm/14) + adx_data["+DM 1"]
    # 8. Formülde ve yazıda yazılan şey farklı. 
    # next_14_plus_dm = first_14_plus_dm - (first_14_plus_dm.shift(1)/14) + first_14_plus_dm
smoothed_plus_dm14()
adx_data[["smoothed_plus_dm14"]]
```

```python
def smoothed_minus_dm14():
    first_14_minus_dm = adx_data["New -DM 1"].rolling(15).sum()
    adx_data["smoothed_minus_dm14"] = first_14_minus_dm - (first_14_minus_dm/14) + adx_data["-DM14"]
    # 8. Formülde ve yazıda yazılan şey farklı. 
    # next_14_minus_dm = first_14_minus_dm - (first_14_minus_dm.shift(1)/14) + first_14_minus_dm
smoothed_minus_dm14()
adx_data[["smoothed_minus_dm14"]]
```

```python
def smoothed_atr_14():
    first_14_atr = adx_data["New TR"].rolling(1).sum()
    adx_data["smoothed_tr"] = first_14_atr - (first_14_atr/14) + adx_data["New TR"]
    # 8. Formülde ve yazıda yazılan şey farklı. 
    # next_14_minus_dm = first_14_minus_dm - (first_14_minus_dm.shift(1)/14) + first_14_minus_dm
smoothed_atr_14()
adx_data[["smoothed_tr"]]
```
