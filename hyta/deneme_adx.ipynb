### -DM KISMI

```python

def minus_dm():
    a = [np.NAN]
    
    for i in range(1,len(adx_data["Low"])):
        low_diff = adx_data["Low"][i-1] - adx_data["Low"][i]
        high_diff = adx_data["High"][i] - adx_data["High"][i-1]
        
        if low_diff > high_diff:
            a.append(max(low_diff,0))
        else:
            a.append(0)
    #return pd.Series({"New -DM": a})
    return pd.Series(a)

minus_dm()
```

### SMOOTHED +DM14, -DM14, ATR
Smooth the 14-period averages of +DM, -DM, and the TR. Below is the formula for TR. <br>
Insert the -DM and +DM values to calculate the smoothed averages of those as well. <br>
First 14TR = Sum of first 14 TR readings. <br>
Next 14TR value = First 14TR - (Prior 14TR/14) + Current TR <br>

```python
def smoothed_plus_dm14():
    first_14_plus_dm = adx_data["+DM 1"][:15].sum()
    islem = first_14_plus_dm - (first_14_plus_dm/14) + adx_data["+DM 1"][15]

    Adx_Data = np.array([])
    for i in range(14):
        Adx_Data = np.append(Adx_Data,np.nan)
    Adx_Data = np.append(Adx_Data,[first_14_plus_dm,islem])
    
    Adx_Data = np.array(Adx_Data)
    for i in range(len(adx_data["+DM 1"])-16):
        value = Adx_Data[-1] - (Adx_Data[-1]/14) + adx_data["+DM 1"][i+16]
        Adx_Data = np.append(Adx_Data,value)
        adx_data["New +DM 14"] = pd.DataFrame({"New +DM 14":Adx_Data})
    return pd.DataFrame(adx_data["New +DM 14"])
a = smoothed_plus_dm14()
a
```

### +DI , -DI

```python
def plus_di_14():
    calculation = 100*(adx_data["New +DM 14"] / adx_data["New Smoothed TR"])
    return calculation

plus_di_14()
```

```python
def minus_di_14():
    calculation = 100*(adx_data["New -DM 14"] / adx_data["New Smoothed TR"])
    return calculation

minus_di_14()
```

```python
def di_diff():
    difference = abs(plus_di_14() - minus_di_14())
    return difference
di_diff()
```

```python
def di_sum():
    sum = plus_di_14() + minus_di_14()
    return sum
di_sum()
```

```python
def dx():
    dx = abs(100*(di_diff() / di_sum()))
    return dx 
dx()
```

```python

### ADX'de kaldım 
def adx():
    first = dx().rolling(13).mean()
    calculation_adx = ((first*13)+dx())/14
    return calculation_adx

adx()
```
